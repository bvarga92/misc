```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mandelbrot(c,maxIterations):
    output=np.empty(c.shape)
    z=np.zeros(c.shape,np.complex64)
    for i in range(maxIterations):
        cmp=np.less(z.real**2+z.imag**2,4.0)
        output[cmp]=i
        z[cmp]=z[cmp]**2+c[cmp]
    output[output==maxIterations-1]=0
    return output
```

```python
def julia(c,maxIterations):
    output=np.empty(c.shape)
    z=c
    for i in range(maxIterations):
        cmp=np.less(z.real**2+z.imag**2,4.0)
        output[cmp]=i
        z[cmp]=z[cmp]**2-0.7+1j*0.27015
    output[output==maxIterations-1]=0
    return output
```

```python
xmin=-2
xmax=1
ymin=-1.5
ymax=1.5
width=500
height=500
maxIterations=25
re=np.linspace(xmin,xmax,width,dtype=np.float32)
im=np.linspace(ymin,ymax,height,dtype=np.float32)
img=mandelbrot(re+1j*im[:,None],maxIterations)
plt.figure(figsize=[8,8]);
plt.imshow(img,extent=[xmin,xmax,ymin,ymax]);
plt.axis([xmin,xmax,ymin,ymax]);
```

```python
xmin=-1.5
xmax=1.5
ymin=-1.5
ymax=1.5
width=500
height=500
maxIterations=500
re=np.linspace(xmin,xmax,width,dtype=np.float32)
im=np.linspace(ymin,ymax,height,dtype=np.float32)
img=julia(re+1j*im[:,None],maxIterations)
plt.figure(figsize=[8,8]);
plt.imshow(img,extent=[xmin,xmax,ymin,ymax]);
plt.axis([xmin,xmax,ymin,ymax]);
```
